```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv("./Downloads/news_data.csv")
```

```python
df.head(2)
```

```python
df_lst = []
for i in tqdm(range(df.__len__())):
    df_ = df['comments.contain_gender_bias.bias.hate'].iloc[i].split(".")
    gen_bi = df_[-3]
    bias = df_[-2]
    hate = df_[-1]
    comments = ""
    for i in df_[:-3]:
        comments += i
    df_lst.append([comments,gen_bi,bias,hate])
    
```

```python
df_lst
```

```python
news = pd.DataFrame(df_lst)
```

```python
news.columns = ['comments','contain_gender_bias','bias','hate']
```

```python
news.to_csv("./news_data_df.csv",index = False, encoding="utf-8")
```
